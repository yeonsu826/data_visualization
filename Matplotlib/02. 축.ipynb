# 02. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.title("꺾은선 그래프",fontdict={'family':'HYGungSo-Bold','size':20}) #개발 폰트 설정 
```

```python
plt.plot(x,y) #어디가 x,y인지 알 수 없음

plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red') # x 축 빨간색
plt.ylabel('Y축', color='#00aa00') # y 축 초록색
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red', loc='right') #left , center, right
plt.ylabel('Y축',color='#00aa00', loc='top') # top, center, bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12]) # 불필요한 tick 간소화
plt.show()
```
