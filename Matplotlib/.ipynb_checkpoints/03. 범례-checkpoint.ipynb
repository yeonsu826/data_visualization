# 03. 범례 (legend)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #widows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시 마이너스 깨짐 현상 해결
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y,label='무슨 데이터')
plt.legend() # 범례 지정
```

```python
plt.plot(x,y,label='무슨 데이터')
plt.legend(loc='upper center') #범례 위치 지정

# right


# upper right
# upper left
# upper center 

# lower left
# lower right
# lower center 

# center
# center left
# center right 
```
