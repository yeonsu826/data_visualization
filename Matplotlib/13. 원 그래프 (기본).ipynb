# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 프로그래밍 언어 선호도 조사
values = [30,25,20,13,10,2]
# values = [1,1,1,1,1,1] 으로 하면 균등하게 나눠 짐
labels = ['Python','Java','Javascirpt','C#','C/C++','ETC']

plt.pie(values, labels = labels, autopct = '%.1f%%') 
# autopct 퍼센트 소수점 첫번째 자리까지 출력 
# %.1f 이렇게 적으면 숫자만 나오지만 
# %.1f%% 이렇게 적으면 숫자% 까지 나옴

plt.show()
```

```python
plt.pie(values, labels = labels, autopct = '%.1f%%', startangle = 90)
# startangle 시작 위치를 90로 부분으로 돌림
plt.show()
```

```python
plt.pie(values, labels = labels, autopct = '%.1f%%', startangle = 90, counterclock = False)
# startangle 시작 위치를 90로 부분으로 돌리고 
# index 순서를 시계방향으로 나열하기 counterclock 을 False로
plt.show()
```

```python
# 강조효과 주기

values = [30,25,20,13,10,2]
# values = [1,1,1,1,1,1] 으로 하면 균등하게 나눠 짐
labels = ['Python','Java','Javascirpt','C#','C/C++','ETC']
# explode = [0.2,0,0,0,0,0]

explode = [0.05] * 6

plt.pie(values, labels=labels, explode = explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode = explode)
plt.legend(loc=(1.2, 0.3)) # 범례 위치 조정
plt.show()
```

```python
plt.pie(values, labels=labels, explode = explode)
plt.title('언어별 선호도 원 그래프') # 언어별 선호도 원 그래프
plt.legend(loc=(1.2, 0.3), title='언어별 선호도 범례')
plt.show()
```
