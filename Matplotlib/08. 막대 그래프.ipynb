# 8. 막대 그래프 기본

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus']= False #한글 폰트 사용시, 마이너스 글자가 깨지는 현상 해결

x=[1,2,3]
y=[2,4,8]
```

```python
labels = ['강백호','서태웅','정대만'] #이름
values = [190,187,184] #키

plt.bar(labels,values)
```

```python
labels = ['강백호','서태웅','정대만'] #이름
values = [190,187,184] #키

plt.bar(labels,values, color='r') #색 변경
```

```python
labels = ['강백호','서태웅','정대만'] #이름
values = [190,187,184] #키
colors = ['r','g','b']

plt.bar(labels,values,color=colors, alpha=0.5) 
# colors배열 입력 색깔을 각 선수마다 다르게
# alpha = 0.5 투명도
```

```python
# y축의 데이터 출력을 제한

labels = ['강백호','서태웅','정대만'] #이름
values = [190,187,184] #키

plt.bar(labels,values)
plt.ylim(175,195) #175~195 y축의 데이터 출력을 제한
```

```python
# 막대 날씬하게 
# width = 0.5

plt.bar(labels,values,width = 0.5)
```

```python
# 각 데이터의 이름이 길 때  
# 각도 바꾸기

plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # x축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y축의 키 데이터 각도를 45도로 설정
```

```python
labels = ['강백호','서태웅','정대만'] #이름
values = [190,187,184] #키
ticks = ['1번 학생','2번 학생','3번 학생']

plt.bar(labels,values, width=0.3)
plt.ylim(175,195) #175~195 y축의 데이터 출력을 제한
plt.xticks(labels,ticks,rotation=30)
```
