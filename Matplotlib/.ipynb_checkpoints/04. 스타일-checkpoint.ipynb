# 4. 스타일

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
plt.show()
```

## 4-1 마커 (Marker)

```python
plt.plot(x,y,linewidth=5) #linewidth 라인 두께
plt.show()
```

```python
plt.plot(x,y,marker='o') # 점 라인
```

```python
plt.plot(x,y,marker='o',linestyle='None') # 선 없이 점만
```

```python
plt.plot(x,y,marker='v',markersize = 10) # 삼각형 마커 marker='v', 마커 크기 조정 markersize
```

```python
plt.plot(x,y,marker='X',markersize=10) #마커 X
```

```python
plt.plot(x,y,marker='$GOOD$',markersize=30) # $_원하는 문자_$
```

### 마커 종류

https://matplotlib.org/stable/api/markers_api.html


```python
plt.plot(x,y,marker='X',markersize=10,markeredgecolor='red')
# markeredgecolor 테두리 색 바꾸기
```

```python
plt.plot(x,y,marker='X', markersize=10,markeredgecolor='red',markerfacecolor='red')
#markerfacecolor 마커 안쪽 색깔 바꾸기 
```

## 4-2 선 스타일

```python
plt.plot(x,y,linestyle=':') # linestyle 점선으로
```

```python
plt.plot(x,y,linestyle='--') #linestyle 좀 더 두껍게
```

```python
plt.plot(x,y,linestyle='-.')
```

### 라인 스타일 

https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html


https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html

- '-'	solid line style 
- '--'	dashed line style
- '-.'	dash-dot line style
- ':'	dotted line style

```python
plt.plot(x,y,linestyle='-')
```

```python
plt.plot(x,y,color='pink') #선 색깔 바꾸기
```

```python
plt.plot(x,y,color='#ff0000') #RGB컬러 설정
```

```python
plt.plot(x,y,color='b') # b : blue
```

### 라인 색깔 정하기 

https://matplotlib.org/stable/gallery/color/named_colors.html



## 4-3 포맷

```python
plt.plot(x,y,'ro--') # color, marker, linestyle
```

```python
plt.plot(x,y,'bv:')
```

```python
plt.plot(x,y,'go') #라인스타일 없을 때 선 없이 나타남
```

```python
plt.plot(x,y,color='g',marker='o',linestyle='None') # 'go' 와 같은 결과
```

## 4-4 축약어

```python
plt.plot(x,y,marker='o',mfc='red',ms=10, mec='blue',ls=':')
# markerfacecolor = mfc 
# makersize = ms
# markeredgecolor = mec
# linestyle = ls 
```

### 축약어
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

## 4-5 투명도

```python
plt.plot(x,y,marker='o',mfc='red',ms=10,alpha=0.3) 
#alpha : 투명도 (0~1) 참고 자료 같은거 넣을 때
```

## 4-6 그래프 크기

```python
plt.figure(figsize=(10,5)) # figsize 튜플형태로 지정 (width , height)
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=50) # dpi : 해상도 (dots fer inch 확대)
plt.plot(x,y)
```

## 4-7 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor='#a1c3ff') # rgb
plt.plot(x,y)
```
