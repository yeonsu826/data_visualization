# 14. 그룹화

동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col ='지원번호')
df # 데이터 가져오기

```

```python
df.groupby('학교') 
```

```python
df.groupby('학교').get_group('북산고')
```

```python
df.groupby('학교').get_group('능남고')
```

```python
df.groupby('학교').mean() #계산 가능한 데이터들의 평균값
```

```python
df.groupby('학교').size()
```
