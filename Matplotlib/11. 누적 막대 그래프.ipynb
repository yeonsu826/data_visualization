# 누적 막대 그래프

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
plt.bar(df['이름'], df['국어'])
plt.bar(df['이름'],df['영어'])
```

```python
plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'],df['영어'], bottom = df['국어']) # 영어성적 위에 국어성적 얹기
```

```python
# plt.bar(df['이름'],df['국어']) # 국어 점수 없다면 영어점수만 표시됨
plt.bar(df['이름'],df['영어'], bottom = df['국어']) # 영어성적 위에 국어성적 얹기
```

```python
plt.bar(df['이름'], df['국어'], label='국어')
plt.bar(df['이름'], df['영어'], bottom = df['국어'], label = '영어') # 영어점수를 위에 표시하기 위해
plt.bar(df['이름'], df['수학'], bottom = df['국어']+df['영어'], label='수학') # 수학 점수를 위에 표시하기 위해

plt.legend(loc='lower right')
plt.xticks(rotation = 60)
```

### 배운 내용

누적 막대 그래프를 그리려면 이전 값들을 밑에 깔아줘야 한다. 
plt.bar(df['국어'], bottom = df['수학'], label='국어') 

표시하련느 값은 국어인데 수학 위에 표시하고 싶다 -> bottom = df['수학'] 으로 한다.
