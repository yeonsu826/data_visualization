# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x = [1,2,3]
y = [2,4,8]

plt.plot(x,y)
```

```python
print(plt.plot(x,y))
```

```python
plt.plot(x,y)

plt.show() # [<matplotlib.lines.Line2D object at 0x000001C051EAE3D0>] 없이 출력
```

## Title 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac

matplotlib.rcParams['font.size'] = 15 # 글자 크기
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist #사용 가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] # font 이름만 출력 
```
